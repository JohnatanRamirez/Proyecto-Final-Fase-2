```python
# Identificar variables numéricas
columnas_numericas = data.select_dtypes(include=np.number).columns

datos_num = data[columnas_numericas]
datos_num

for columna_actual in columnas_numericas[datos_num.isnull().any()]:

    # Imputación CCA:
    datos_num_CCA = datos_num.dropna(subset=[columna_actual])
    registros_eliminados_CCA = datos_num.shape[0] - datos_num_CCA.shape[0]
    distribucion_CCA = datos_num_CCA[columna_actual].describe()

    # Imputación de media
    datos_num_media = datos_num.fillna(datos_num[columna_actual].mean())
    distribucion_media = datos_num_media[columna_actual].describe()

    # Imputación de mediana
    datos_num_mediana = datos_num.fillna(datos_num[columna_actual].median())
    distribucion_mediana = datos_num_mediana[columna_actual].describe()

    # Imprimir resultados para la columna actual
    print(f"\n**Columna:** {columna_actual}")
    print("-------------------------")
    print("Análisis CCA:")
    print(f"Registros eliminados: {registros_eliminados_CCA}")
    print(distribucion_CCA)
    print("\nAnálisis Imputación Media:")
    print(distribucion_media)
    print("\nAnálisis Imputación Mediana:")
    print(distribucion_mediana)
```

### Tratamiento de Outliers por Capping

```python
for colName in continuas:

    IQR = data2[colName].quantile(0.75) - data2[colName].quantile(0.25)
    LI = data2[colName].quantile(0.25) - 1.5*IQR
    LS = data2[colName].quantile(0.25) + 1.5*IQR

    data2[colName] = np.where(data2[colName] > LS, LS,
                             np.where(data2[colName] < LI, LI, data2[colName]))
```
